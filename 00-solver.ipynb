```python
import pandas as pd
import numpy as np
from ortools.algorithms import pywrapknapsack_solver
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)
```

```python
df = pd.read_pickle('objective.pkl')
```

```python
start_gw = 1 #what's the gameweek to begin optimizing from?
num_gws = 5 #for this example, going to optimize transfers for five gameweeks
objective_column = 'fwd_5_pts'
week = 1
data = df[df.gw == week]
data = data.groupby(['name', 'position', 'team', 'value', 'element'])[objective_column].mean().reset_index() #group, since there are some double gameweeks
```

```python
solver = pywrapknapsack_solver.KnapsackSolver(
    pywrapknapsack_solver.KnapsackSolver.
    KNAPSACK_MULTIDIMENSION_BRANCH_AND_BOUND_SOLVER, 'KnapsackExample')
```

```python
np.ones(len(data.value)).astype('int')
```

```python
#to add additional weights, add additional lists to the weights lists
#to add additional capacities, add additional capicities to the capicities lists

values = list(data.fwd_5_pts.values)
weights = [list(data.value.values), list(np.ones(len(data.value)).astype('int'))]
capacities = [100*10, 11]

solver.Init(values, weights, capacities)
computed_value = solver.Solve()
packed_items = []
packed_weights = []
total_weight = 0
print('Total value =', computed_value)
for i in range(len(values)):
    if solver.BestSolutionContains(i):
        packed_items.append(i)
        packed_weights.append(weights[0][i])
        total_weight += weights[0][i]
print('Total weight:', total_weight)
print('Packed items:', packed_items)
print('Packed_weights:', packed_weights)
```
